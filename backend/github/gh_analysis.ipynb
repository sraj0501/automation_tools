```python
from datetime import datetime, timedelta

def convert_to_ist(date_val, date_format="%Y-%m-%dT%H:%M:%SZ"):
    utc_time = datetime.strptime(date_val, date_format)
    ist_offset = timedelta(hours=5, minutes=30)
    ist_time = utc_time + ist_offset
    return ist_time.date()
```

```python
def get_commit_details(owner, repo, sha, token):
    url = f"https://api.github.com/repos/{owner}/{repo}/commits/{sha}"
    headers = {"Authorization": f"Bearer {token}"}

    response = requests.get(url, headers=headers)
    commit_data = response.json()

    if response.status_code != 200:
        print(f"Error fetching commit details: {response.status_code}")
        return None

    # Extract relevant commit information
    commit_info = {
        "sha": commit_data['sha'],
        "author": commit_data['commit']['author']['name'],
        "date": commit_data['commit']['author']['date'],
        "message": commit_data['commit']['message'],
        "files": [file['filename'] for file in commit_data.get('files', [])],
        "total_changes": sum(file['changes'] for file in commit_data.get('files', [])),
        "file_changes": [{"file": file['filename'], "additions": file['additions'], "deletions": file['deletions'], "changes": file['changes']} for file in commit_data.get('files', [])]
    }
    return commit_info
```

```python
import requests
from datetime import datetime, timedelta

def get_all_commits(owner, repo, token):
    # Start with the basic URL for fetching commits
    url = f"https://api.github.com/repos/{owner}/{repo}/commits"
    headers = {"Authorization": f"Bearer {token}"}

    all_commits = []
    
    while True:
        response = requests.get(url, headers=headers)
        commits = response.json()

        # Debugging: Print fetched commits and response status
        print(f"Response status: {response.status_code}")
        print(f"Commits fetched: {len(commits)}")

        # Check for errors in response
        if response.status_code != 200:
            print(f"Error fetching commits: {response.status_code}")
            return []

        # Extend the list of all commits
        all_commits.extend(commits)

        # Handle pagination by checking the 'Link' header for a next page
        next_link = response.headers.get("Link", None)
        if next_link:
            # Extract the next URL from the Link header
            if 'rel="next"' in next_link:
                url = next_link.split(",")[0].split(";")[0].strip()[1:-1]
                print(f"Next page link: {url}")
            else:
                break  # No more pages
        else:
            break  # No Link header, so stop fetching

    return all_commits

def get_commit_details(owner, repo, sha, token):
    url = f"https://api.github.com/repos/{owner}/{repo}/commits/{sha}"
    headers = {"Authorization": f"Bearer {token}"}

    response = requests.get(url, headers=headers)
    commit_data = response.json()

    if response.status_code != 200:
        print(f"Error fetching commit details: {response.status_code}")
        return None

    # Extract relevant commit information
    commit_info = {
        "sha": commit_data['sha'],
        "author": commit_data['commit']['author']['name'],
        "date": commit_data['commit']['author']['date'],
        "message": commit_data['commit']['message'],
        "files": [file['filename'] for file in commit_data.get('files', [])],
        "total_changes": sum(file['changes'] for file in commit_data.get('files', [])),
        "file_changes": [
            {"file": file['filename'], "additions": file['additions'], "deletions": file['deletions'], "changes": file['changes']}
            for file in commit_data.get('files', [])
        ]
    }
    return commit_info


commits = get_all_commits(owner, repo, token)
for commit in commits:
    commit_details = get_commit_details(owner, repo, commit['sha'], token)
    if commit_details:
        print(f"Commit by {commit_details['author']} on {commit_details['date']}: {commit_details['message']}")
        print(f"Files changed: {commit_details['files']}")
        print(f"Total changes: {commit_details['total_changes']}")
        print(f"Details: {commit_details['file_changes']}")
        print("=" * 50)
```
